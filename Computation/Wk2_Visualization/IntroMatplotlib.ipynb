```python
import numpy as np

def make_vars(x):
    numbers = np.random.normal(size = (x,x))
    arith_m = numbers.mean(axis=1)
    return np.var(arith_m)

def wlln(n):
    vars = np.zeros(10)
    for i in range(10):
        vars[i] = make_vars(n[i])
    return vars
```

```python
import matplotlib.pyplot as plt

n = np.linspace(100, 1000, 10, dtype=int)
v = wlln(n)
plt.plot(n,v)
plt.show()
```

```python
def plot_trig():
    x = np.linspace(-2*np.pi, 2*np.pi, 1000)
    s = np.sin(x)
    c = np.cos(x)
    ar = np.arctan(x)
    plt.plot(x, s)
    plt.plot(x, c)
    plt.plot(x, ar)
    plt.show()
```

```python
plot_trig()
```

### Problem 3

```python
def plot_curve(l, r):
    x1 = np.linspace(l, 0.99, abs(l)*10)
    x2 = np.linspace(1.01, r, abs(r)*10)
    f1 = 1 / (x1 -1)
    f2 = 1 / (x2 -1)
    plt.plot(x1, f1, '--', lw=3, color='m')
    plt.plot(x2, f2, '--', lw=3, color='m')
    plt.ylim(-6, 6)
    plt.show()
```

```python
plot_curve(l=-2, r=6)
```

### Problem 4.

```python
def more_trig():
    x = np.linspace(0, 2*np.pi)
    s = np.sin(x)
    s2 = np.sin(2*x)
    ss = 2*s
    ss2 = 2*s2
    ys = [s, s2, ss, ss2]
    tits = ['sin(x)', 'sin(2x)','2sin(x)', '2sin(2x)']
    cols = ['g', 'r', 'b', 'm']
    style = ['-', '--', '--', ':']
    fignum = range(1, 5)
    for i in fignum:
        plt.subplot(2, 2, i)
        plt.plot(x, ys[i-1], style[i-1], color=cols[i-1])
        plt.axis([0, 2*np.pi, -2, 2])
        plt.title(tits[i-1])
    plt.suptitle('more trigonometric functions')
    plt.show()
    
```

```python
more_trig()
```

### Problem 5.

```python
fars = np.load(r'FARS.npy')
def plot_fars():
    plt.subplot(2,1,1)
    long = fars[:,-2]
    lat = fars[:, -1]
    plt.plot(long, lat, 'k,')
    plt.axis('equal')
    plt.ylabel('longitude')
    plt.xlabel('latitude')
    
    plt.subplot(2,1,2)
    hrs = fars[:,0]
    plt.hist(hrs, bins=24, range=[0,23])
    
    plt.tight_layout()
    plt.show()
```

```python
plot_fars()
```

### Problem 6.

```python
def plot_combi_trigs():
    x = np.linspace(-2*np.pi, 2*np.pi)
    y = x.copy()
    X, Y = np.meshgrid(x, y)
    f = np.sin(X)*np.sin(Y) / (X * Y)
    
    plt.subplot(221)
    plt.axis([2*-np.pi, 2*np.pi,2*-np.pi, 2*np.pi])
    plt.pcolormesh(X, Y, f, cmap='RdBu_r')
    plt.colorbar()
    
    plt.subplot(222)
    plt.contour(X, Y, f, np.linspace(-0.2, 0.8, 10), cmap='RdBu_r')
    plt.axis([2*-np.pi, 2*np.pi,2*-np.pi, 2*np.pi])
    plt.colorbar()
    
    plt.show()
    
```

```python
plot_combi_trigs()
```
